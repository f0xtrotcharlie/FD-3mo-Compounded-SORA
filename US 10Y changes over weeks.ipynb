```python
import matplotlib.pyplot as plt
import pandas as pd
import quandl as ql
#import Quandl as ql
%matplotlib inline
```

```python
yield_ = ql.get("USTREASURY/YIELD")

today = yield_.iloc[-7,:]
yest = yield_.iloc[-14,:]
#fn = yield_.iloc[-15,:]
month_ago = yield_.iloc[-30,:]

df = pd.concat([today, yest, month_ago], axis=1)

#df = pd.concat([today,yest, fn, month_ago], axis=1)

df.columns = ['Today','Yest', 'month ago']

df.plot(style={'Today': 'go-', 'Yest': 'ro-',  'month ago': 'bx--'}
        ,title='Treasury Yield Curve, %');
        
#df.columns = ['Today','Yest','2 weeks ago', 'month ago']
              
#df.plot(style={'Today': 'go-', 'Yest': 'ro-', '2 weeks ago': 'bo-', 'month ago': 'bx--'}
#        ,title='Treasury Yield Curve, %');
```
